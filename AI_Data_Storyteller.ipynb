````python
eda_code = """
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os

# =====================
# Dataset Handling
# =====================

def load_dataset(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except Exception as e:
        return f"Error loading dataset: {e}"

def dataset_summary(df):
    summary = {
        "shape": df.shape,
        "columns": df.columns.tolist(),
        "missing_values": df.isnull().sum().to_dict(),
        "data_types": df.dtypes.astype(str).to_dict(),
        "descriptive_stats": df.describe(include='all').to_dict()
    }
    return summary

# =====================
# Visualization Functions
# =====================

def plot_histograms(df, output_dir="eda_charts"):
    os.makedirs(output_dir, exist_ok=True)
    for col in df.select_dtypes(include=['int64','float64']).columns:
        plt.figure()
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f"Histogram of {col}")
        plt.savefig(f"{output_dir}/hist_{col}.png")
        plt.close()

def plot_categorical(df, output_dir="eda_charts"):
    os.makedirs(output_dir, exist_ok=True)
    for col in df.select_dtypes(include=['object','category']).columns:
        plt.figure(figsize=(6,4))
        sns.countplot(y=df[col], order=df[col].value_counts().index)
        plt.title(f"Countplot of {col}")
        plt.savefig(f"{output_dir}/count_{col}.png")
        plt.close()

def plot_correlation(df, output_dir="eda_charts"):
    os.makedirs(output_dir, exist_ok=True)
    plt.figure(figsize=(8,6))
    corr = df.corr(numeric_only=True)
    sns.heatmap(corr, annot=True, cmap="coolwarm")
    plt.title("Correlation Heatmap")
    plt.savefig(f"{output_dir}/correlation_heatmap.png")
    plt.close()

def plot_boxplots(df, output_dir="eda_charts"):
    os.makedirs(output_dir, exist_ok=True)
    for col in df.select_dtypes(include=['int64','float64']).columns:
        plt.figure()
        sns.boxplot(x=df[col].dropna())
        plt.title(f"Boxplot of {col}")
        plt.savefig(f"{output_dir}/box_{col}.png")
        plt.close()

# =====================
# Combined EDA Pipeline
# =====================

def run_auto_eda(df, output_dir="eda_charts"):
    os.makedirs(output_dir, exist_ok=True)
    plot_histograms(df, output_dir)
    plot_categorical(df, output_dir)
    plot_correlation(df, output_dir)
    plot_boxplots(df, output_dir)
    return f"EDA completed. Charts saved in {output_dir}/"
"""

with open("/content/AI_Data_Storyteller/storyteller_core.py", "w") as f:
    f.write(eda_code.strip())

print("✅ storyteller_core.py updated with EDA functions.")
````

````python
ai_code = """
import json

# ==============
# AI Integration
# ==============

# Prompt Template
def build_prompt(summary, insights=""):
    prompt = f'''
    You are a professional Data Analyst.
    Based on the dataset summary and statistics below, write a clear,
    business-style analysis in 3–5 paragraphs.
    Highlight key patterns, correlations, anomalies, and possible business implications.

    Dataset Summary:
    {json.dumps(summary, indent=2)}

    Additional Insights:
    {insights}

    Write the explanation in plain English (non-technical).
    '''
    return prompt

# OpenAI API Integration
def generate_ai_narrative_openai(summary, insights="", model="gpt-4o-mini"):
    from openai import OpenAI
    client = OpenAI()

    prompt = build_prompt(summary, insights)

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are an expert data analyst."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500,
        temperature=0.7
    )
    return response.choices[0].message.content

# Ollama Local LLM Integration
def generate_ai_narrative_ollama(summary, insights="", model="llama3"):
    import subprocess, json

    prompt = build_prompt(summary, insights)

    result = subprocess.run(
        ["ollama", "run", model],
        input=prompt.encode("utf-8"),
        capture_output=True,
        text=True
    )
    return result.stdout
"""

with open("/content/AI_Data_Storyteller/storyteller_core.py", "a") as f:
    f.write("\n" + ai_code.strip())

print("✅ storyteller_core.py updated with AI functions")
````

````python
!pip -q install reportlab python-docx

from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
from docx import Document
from docx.shared import Inches
import glob, textwrap, os

project_root = "/content/AI_Data_Storyteller"
chart_dir = f"{project_root}/eda_charts"
md_path = f"{project_root}/report.md"

# --- PDF Export ---
def export_pdf(md_text, chart_dir, out_path=f"{project_root}/report.pdf"):
    c = canvas.Canvas(out_path, pagesize=A4)
    width, height = A4
    margin = 40
    y = height - margin

    # Title
    c.setFont("Helvetica-Bold", 16)
    c.drawString(margin, y, "AI Data Storyteller Report")
    y -= 24

    c.setFont("Helvetica", 10)
    for line in md_text.splitlines():
        # crude markdown to text; wrap long lines
        line = line.replace("## ", "").replace("# ", "")
        wrapped = textwrap.wrap(line, width=95)
        for w in wrapped:
            if y < 100:
                c.showPage(); y = height - margin; c.setFont("Helvetica", 10)
            c.drawString(margin, y, w)
            y -= 14
        if line.strip() == "":
            y -= 6

    # add charts
    chart_files = sorted(glob.glob(os.path.join(chart_dir, "*.png")))
    for img in chart_files:
        if y < 250:
            c.showPage(); y = height - margin
        try:
            c.drawImage(ImageReader(img), margin, y-220, width=width-2*margin, height=220, preserveAspectRatio=True, anchor='n')
            y -= 240
        except Exception:
            pass

    c.save()
    return out_path

# --- DOCX Export ---
def export_docx(md_text, chart_dir, out_path=f"{project_root}/report.docx"):
    doc = Document()
    doc.add_heading("AI Data Storyteller Report", level=1)
    for line in md_text.splitlines():
        if line.startswith("## "):
            doc.add_heading(line[3:], level=2)
        elif line.startswith("# "):
            doc.add_heading(line[2:], level=1)
        else:
            doc.add_paragraph(line)

    chart_files = sorted(glob.glob(os.path.join(chart_dir, "*.png")))
    for img in chart_files:
        doc.add_picture(img, width=Inches(5.8))
    doc.save(out_path)
    return out_path

# Load narrative text
with open(md_path, "r", encoding="utf-8") as f:
    md_text = f.read()

pdf_path = export_pdf(md_text, chart_dir)
docx_path = export_docx(md_text, chart_dir)

print("✅ PDF:", pdf_path)
print("✅ DOCX:", docx_path)
````

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
import openai, os
import gradio as gr

# =============== 1. Dataset summary =================
def dataset_summary(df):
    summary = {
        "Shape": df.shape,
        "Columns": list(df.columns),
        "Missing Values": df.isnull().sum().to_dict(),
        "Numeric Summary": df.describe(include="number").to_dict()
    }
    return summary

# =============== 2. Auto EDA charts =================
def run_auto_eda(df, out_file="eda_report.html"):
    profile = ProfileReport(df, explorative=True)
    profile.to_file(out_file)
    return out_file

# =============== 3. AI narrative via OpenAI =================
openai.api_key = "YOUR_OPENAI_API_KEY"   # 🔑 replace

def generate_ai_narrative_openai(summary_text):
    prompt = f"""
    You are a data analyst. Create a consulting-style structured report with these sections:
    - Key Stats
    - Trends & Insights
    - Anomalies
    - Business Implications

    Dataset Summary:
    {summary_text}
    """
    try:
        response = openai.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role":"user","content":prompt}]
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"⚠️ Error: {str(e)}"
````

````python
def pipeline(file):
    # Load CSV
    df = pd.read_csv(file.name)

    # Dataset summary
    summary = dataset_summary(df)
    summary_text = str(summary)

    # Auto EDA HTML report
    eda_file = run_auto_eda(df)
    eda_link = f"<a href='file/{eda_file}' target='_blank'>📊 View EDA Report</a>"

    # AI Narrative
    ai_text = generate_ai_narrative_openai(summary_text)

    return str(summary), eda_link, ai_text


# =============== Gradio Interface =================
with gr.Blocks() as demo:
    gr.Markdown("## 📈 AI + Data Science Hybrid Tool (Gradio Edition)")
    gr.Markdown("Upload a dataset → Auto Summary + EDA + AI Narrative Report")

    with gr.Row():
        file = gr.File(label="Upload CSV", file_types=[".csv"])

    with gr.Row():
        summary_out = gr.Textbox(label="📑 Dataset Summary")

    with gr.Row():
        eda_out = gr.HTML(label="📊 EDA Report (HTML Link)")

    with gr.Row():
        ai_out = gr.Textbox(label="🤖 AI Narrative Report", lines=15)

    file.change(fn=pipeline, inputs=file, outputs=[summary_out, eda_out, ai_out])

demo.launch(share=True)
````
